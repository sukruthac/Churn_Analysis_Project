```python
#check the unque valus for 'Churn'-here True-customer left from company,False-Customer active
tel_data['Churn'].unique()
```

```python
#display all the 'churn' customer with false value
tel_data[tel_data['Churn'] == False]
```

```python
#convert churn into numeric data
tel_data['Churn_numeric'] = tel_data['Churn'].map({True: 1, False: 0})
tel_data['Churn_numeric'] 
```

```python
#false churn and true churn data
false_churn = tel_data[(tel_data['Churn'])==False]
True_churn = tel_data[(tel_data['Churn'])==True]
print(false_churn,True_churn)
print("false-count",false_churn.size)
print("True-count",True_churn.size)
```

```python
#find churn percentage-percentage of customers left the telecom service
churn_percentage = tel_data['Churn_numeric'].mean() * 100
churn_percentage
```

```python
#compare with Total night charge
tel_data.groupby('Churn')['Total night charge'].mean()
```

```python
#Total intl charge
tel_data.groupby('Churn')['Total intl charge'].mean()
```

```python
#Total eve charge
tel_data.groupby('Churn')['Total eve charge'].mean()
```

```python
#Total day charge
tel_data.groupby('Churn')['Total day charge'].mean()
```
